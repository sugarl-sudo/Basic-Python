```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

```python
df_task03 = pd.read_csv("../quizzes/matplotlib_quiz/line_data.csv")
fig, ax = plt.subplots()
ax.plot(df_task03['期間'], df_task03['件数'], marker="o")
xlabels = ax.get_xticklabels()
ax.set_title('期間ごとの件数の推移')
plt.setp(xlabels, rotation=30)
ax.set_xlabel('期間')
ax.set_ylabel('件数')
ax.grid()
plt.show()
```

```python
df_task03 = pd.read_csv("../quizzes/matplotlib_quiz/line_data.csv")
fig, ax = plt.subplots()
ax.bar(df_task03['期間'], df_task03['件数'])
xlabels = ax.get_xticklabels()
ax.set_title('期間ごとの件数の推移')
plt.setp(xlabels, rotation=30)
ax.set_xlabel('期間')
ax.set_ylabel('費用(円)')
ax.grid()
plt.show()
```
