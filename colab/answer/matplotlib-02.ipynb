```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
```

```python
df_task02 = pd.read_csv("../quizzes/matplotlib_quiz/scatter_data.csv")
fig, ax = plt.subplots()
for name, group in df_task02.groupby('class'):
    ax.scatter(group["length (cm)"], group["width (cm)"], label="class"+str(name))

ax.set_title('クラス0～2のlengthとwidthに関する散布図')
ax.set_xlabel('length（cm）')
ax.set_ylabel('width（cm）')
ax.grid(True)
leg = ax.legend()
plt.show()
```
