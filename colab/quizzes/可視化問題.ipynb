```python
df_task02 = pd.read_csv("matplotlib_quiz/scatter_data.csv")
df_task02.head()
```

```python
df_task02 = pd.read_csv("matplotlib_quiz/scatter_data.csv")
fig, ax = plt.subplots()
for name, group in df_task02.groupby('class'):
    ax.scatter(group["length (cm)"], group["width (cm)"], label="class"+str(name))

ax.set_title('クラス0～2のlengthとwidthに関する散布図')
ax.set_xlabel('length（cm）')
ax.set_ylabel('width（cm）')
ax.grid(True)
leg = ax.legend()
plt.show()
```

```python
df_task03 = pd.read_csv("matplotlib_quiz/line_data.csv")
df_task03.head()
```

```python
fig, ax = plt.subplots()
ax.plot(df_task03['期間'], df_task03['件数'], marker="o")
ax.set_title('期間ごとの件数の推移')
ax.set_xlabel('期間')
fig.autofmt_xdate(rotation=30)
ax.set_ylabel('件数')
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df_task03['期間'], df_task03['費用'])
ax.set_title('期間ごとの費用の推移')
ax.set_xlabel('期間')
fig.autofmt_xdate(rotation=30)
ax.set_ylabel('費用(円)')
ax.grid()
plt.show()
```
